Adcionar nomes a um arquivo.

```python
def adicionar_nome(nome):
    with open('nomes.txt', 'a') as file:
        file.write(nome + '\n')


while True:
    nome = input("Digite um nome (ou 'sair' para terminar): ")
    if nome.lower() == 'sair':
        break
    adicionar_nome(nome)
```

Criar um arquivo json com o nome e telefone.

```python
import json

# Função para adicionar um contato ao arquivo
def adicionar_contato(nome, telefone):
    contato = {"nome": nome, "telefone": telefone}
    with open('contatos.json', 'a') as file:
        json.dump(contato, file)
        file.write('\n')

# Laço para inserir contatos
while True:
    nome = input("Digite o nome do contato (ou 'sair' para terminar): ")
    if nome.lower() == 'sair':
        break
    telefone = input("Digite o telefone do contato: ")
    adicionar_contato(nome, telefone)
```

Exercício 1


Você é responsável por registrar as tarefas diárias em um arquivo. Crie um programa que permita inserir uma tarefa e salve essas tarefas em um arquivo. O programa deve parar de registrar quando o usuário digitar "sair".

Instruções:
Implemente um script que:
 - Peça ao usuário para digitar uma tarefa.
 - Adicione a tarefa ao arquivo tarefas.txt.
 - Termine quando o usuário digitar "sair".

image.png

image.png

```python
# Criar uma função que escreve a tarefa dentro do arquivo
def adicionar_tarefa(tarefa):
  with open('tarefas.txt', 'a') as file:
    file.write(tarefa + '\n')

# Criar um programa que colha uma tarefa do usuario e chame a função de escrever tarefa
while True:
  tarefa = input('Digite uma tarefa ou sair para encerrar: ')
  if tarefa == 'sair':
    break
  else:
    adicionar_tarefa(tarefa)
```

Exercício 2

Você é responsável por criar um sistema de biblioteca que permita aos usuários fazer login e consultar se um determinado livro está disponível. Primeiro, o usuário deve fazer login com nome de usuário e senha, e depois, pode consultar se um livro está na lista de livros disponíveis.

Instruções:
 - Crie um arquivo usuarios.json com os dados de usuários:
 - Crie um arquivo livros.txt com uma lista de livros disponíveis, cada um em uma linha:
 - Implemente um script que:

    - Peça ao usuário para fazer login.
    - Se o login for bem-sucedido, permita ao usuário consultar se um livro está disponível.
    - Termine quando o usuário digitar "sair" no menu de consulta.

image.png

```python
import json
import getpass

# Função para carregar os logins
def carrega_login():
  with open('/content/usuarios.txt', 'r') as file:
    var = json.load(file)
  return var

# Função para carregar os livros
def carrega_livros():
  with open('/content/livros.txt', 'r') as file:
    return file.read().splitlines()

# Função Verificar o login e senha
def verifica_login(usuarios, login, senha):
    for i in usuarios:
      if i['username'] == login and i['password'] == senha:
        return True
    return False

# Função verificar os livros
def verifica_livro(livros, livro):
  return livro in livros

# Programa que faça login e verifique livros

# Carregar os usuarios e a lista de livros
usuarios = carrega_login()
livros = carrega_livros()

# Laço: Para verificar o login
logado = False
while not logado:
  login = input('Digite o login: ')
  senha = getpass.getpass('Digite a senha: ')
  if verifica_login(usuarios, login, senha):
    print('Login bem sucedido!')
    logado = True
  else:
    print('Login mal sucedido, tente novamente!')

while True:
  livro = input('Digite o nome do livro que deseja verificar, ou sair para encerrar: ')
  if livro == 'sair':
    print('Encerrando.')
    break
  else:
    if verifica_livro(livros, livro):
      print(f'O livro {livro} está disponível!')
    else:
      print(f'O livro {livro} não está disponível!')
```
